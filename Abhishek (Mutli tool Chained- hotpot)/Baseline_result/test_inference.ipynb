```python
import pandas as pd
from fuzzywuzzy import fuzz

def exact_match(text1, text2):
    return text1 == text2

def fuzzy_match(text1, text2):
    ratio = fuzz.partial_ratio(text1.lower(), text2.lower())
    return ratio

def check_answer(row):
    gen = str(row['untrained_Gemma2b_ans'])
    act = str(row['A'])
    if exact_match(gen, act):
        return True
    if gen.find(act) != -1:
        return True
    if gen == act:
        return True
    if fuzzy_match(gen, act) > 70:  # You can adjust the threshold for fuzzy matching
        return True
    return False

# Assuming df is your DataFrame
def apply_comparison(df):
    df['Comparison'] = df.apply(check_answer, axis=1)
    return df

# Example of usage with a DataFrame named df
# df = pd.DataFrame({
#     'untrained_Mystral7b_ans': ['example answer', 'another answer'],
#     'A': ['example answer', 'different answer']
# })

# df = apply_comparison(df)
# print(df)

df = apply_comparison(df)
df
```
